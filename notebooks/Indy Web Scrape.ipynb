```python
def domaincounts(data):
    if 'domain' in data.columns:
        return data.groupby('domain').size().sort_values(ascending=False)
    else: 
        data['domain'] = data.url.apply(domain).apply(stripwww)
        return data.groupby('domain').size().sort_values(ascending=False)
def domain(url):
    parsed_url = urllib.parse.urlparse(url)
    netloc = parsed_url.netloc
    return netloc
def stripwww(url):
    if 'www' in url:
        url.split('.')
        url = '.'.join(url.split('.')[1:])
        return url.lower()
    else:
        return url.lower()
def removewww(url):
    if 'www' in url:
        url.split('.')
        url = '.'.join(url.split('.')[1:])
        return url.lower()
    elif 'http' in url:
        url.split('/')
        url = '/'.join(url.split('/')[2:])
        return url.lower()    
def searchtitle(search_term, data):
    return list(data[data.title.str.contains(search_term)].url.sort_values())
```
